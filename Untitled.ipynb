```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

```

```python
def load_data(filepath):
   
    data = pd.read_csv(filepath)
    
 
    features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 
                'waterfront', 'view', 'condition', 'grade', 'sqft_above', 
                'sqft_basement', 'yr_built', 'yr_renovated', 'lat', 'long']
    target = 'price'
    
   
    data = data.fillna(data.mean(numeric_only=True))
    
  
    X = data[features].values
    y = data[target].values
    
    return data, X, y, features

def preprocess_data(X):
   
   
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    std[std == 0] = 1
    
   
    X_norm = (X - mean) / std
    
   
    m = X_norm.shape[0]
    ones = np.ones((m, 1))
    X_ready = np.concatenate((ones, X_norm), axis=1)
    
    return X_ready

def train_test_split_custom(X, y, test_ratio=0.2, seed=42):
   
    np.random.seed(seed) 
    indices = np.random.permutation(X.shape[0])
    test_size = int(X.shape[0] * test_ratio)
    
    test_indices = indices[:test_size]
    train_indices = indices[test_size:]
    
    X_train, X_test = X[train_indices], X[test_indices]
    y_train, y_test = y[train_indices], y[test_indices]
    
    return X_train, X_test, y_train, y_test
```

```python
def normal_equation(X, y):
   
    #Transpose X
    X_T = X.T
    
    #Compute X Transpose t
    XT_X = np.dot(X_T, X)
    
    #Compute Inverse
    epsilon = 1e-5 * np.eye(XT_X.shape[0]) 
    XT_X_inv = np.linalg.inv(XT_X + epsilon)
    
    # Multiply Inverse by X Transpose
    XT_X_inv_XT = np.dot(XT_X_inv, X_T)
    
    #Final Weights
    theta = np.dot(XT_X_inv_XT, y)
    
    return theta

def predict(X, theta):
   
    return np.dot(X, theta)
```

```python
def evaluate(y_true, y_pred):
   
    mse = np.mean((y_true - y_pred)**2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(y_true - y_pred))
    
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    r2 = 1 - (ss_res / ss_tot)
    
    return mse, rmse, mae, r2

def plot_correlation_matrix(df, features, target):
   
    print("   > Drawing Correlation Matrix...")
    
    
    subset_data = df[features + [target]]
    
    
    corr_matrix = subset_data.corr()
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
    plt.title('Feature Correlation Matrix')
    plt.show()

def plot_residuals(y_test, y_pred):
  
    print("   > Drawing Residual Plot...")
    residuals = y_test - y_pred
    
    plt.figure(figsize=(10, 6))
    plt.scatter(y_pred, residuals, alpha=0.5, color='green')
    plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
    plt.xlabel('Predicted Prices')
    plt.ylabel('Residuals (Error)')
    plt.title('Residual Plot: Are our errors random?')
    plt.grid(True, alpha=0.3)
    plt.show()

def analyze_dimensionality(X, y):

    print("   > Running Dimensionality Analysis (checking speed)...")
    feature_counts = range(1, X.shape[1] + 1, 2)
    times = []
    
    for count in feature_counts:
        X_sub = X[:, :count]
        start_time = time.time()
        
        
        if 'normal_equation' in globals():
            _ = normal_equation(X_sub, y)
        else:
            print("Error: Run Cell 3 first!")
            return
            
        end_time = time.time()
        times.append(end_time - start_time)
        
    plt.figure(figsize=(10, 5))
    plt.plot(feature_counts, times, marker='o', linestyle='-', color='purple')
    plt.title('Impact of Data Dimensionality on Speed')
    plt.xlabel('Number of Features')
    plt.ylabel('Time (seconds)')
    plt.grid(True)
    plt.show()
```

```python
print("Loading the housing data")
df_raw, X_raw, y_raw, feature_names = load_data('Housing.csv')

print("Generating Correlation Matrix...")
plot_correlation_matrix(df_raw, feature_names, 'price')       

print(" Preprocessing (Scaling values)")
X_processed = preprocess_data(X_raw)

print("Splitting data")
X_train, X_test, y_train, y_test = train_test_split_custom(X_processed, y_raw)
print(f"   - Training on {X_train.shape[0]} samples")
print(f"   - Testing on  {X_test.shape[0]} samples")

print(" Training the model")
start_time = time.time()
theta = normal_equation(X_train, y_train)
end_time = time.time()
print(f"   - Done! Model trained in {end_time - start_time:.5f} seconds.")

print("Evaluating performance")
y_pred = predict(X_test, theta)
mse, rmse, mae, r2 = evaluate(y_test, y_pred)

print("-" * 40)
print(f"FINAL REPORT CARD:")
print(f"Accuracy (R-Squared): {r2:.4f}")
print(f"RMSE (Avg Error):     {rmse:,.2f}")
print(f"MAE  (Mean Error):    {mae:,.2f}")
print("-" * 40)
```

```python
print("Generating Final Graphs")

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='blue', label='Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2, label='Perfect Fit')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Price')
plt.title('Reality vs. Prediction')
plt.legend()
plt.show()

plot_residuals(y_test, y_pred)

analyze_dimensionality(X_processed, y_raw)
```
